    Write code that prints Hello if 1 is stored in spam, prints Howdy if 2 is stored in spam, and prints Greetings! if anything else is stored in spam.

```python
spam = int(input("Enter a number %% 0 and 3: "))
```

```python
if spam == 1:
  print("Hello")
if spam == 2:
  print("Howdy")
else:
  print("Greetings!")
```
