Null Values

```python
data = {
'Name': ['Ama', 'Kojo', 'Yaw', 'Esi', 'Kofi'],
'Age': [25, None, 30, None, 40],
'City': ['Accra', 'Kumasi', None, 'Takoradi', None],
'Salary': [5000, 4200, None, 3900, None]
}
```

```python
# 1. Find how many null values are in each column.
```

```python
Work = pd.DataFrame(data)
Work
```

```python
Work.isnull().sum()
```

```python
 #2. Select only the rows where City is null.
```

```python
Work[Work['City'].isnull()]
```

```python
# 3. Select only the rows where Salary is not null.
```

```python
Work[Work['Salary'].notnull()]
```

```python
# 4. Drop all rows that contain ANY null values.
```

```python
Work.dropna()
```

```python
# 5. Drop only the rows where Age is null.
```

```python
Work.dropna(subset=['Age'])
```

```python
# 6. Fill null ages with the mean Age
```

```python
Work['Age'] = Work['Age'].fillna(Work['Age'].mean())
Work
```

```python
# 7. Fill null cities with "Unknown"
```

```python
Work['City'] = Work['City'].fillna('Unknown')
Work
```

```python
# 8. Fill null salaries with the median Salary
```

```python
Work['Salary'] = Work['Salary'].fillna(Work['Salary'].median())
Work
```

```python
# 9. Use forward fill (ffill) to fill all null values.
```

```python
Work = Work.fillna(method = 'ffill')
```

```python
# 10. Use backward fill (bfill) to fill all null values.
```

```python
Work = Work.fillna(method = 'bfill')
```
